```python
import pandas as pd

d = pd.read_csv("covid_19_india.csv")
d.head()

d["Date"] = pd.to_datetime(d["Date"])
d.columns=d.columns.str.replace("/","_").str.lower()
d.sort_values(by=["state_unionterritory", "date"])
d.head(10)

```

```python
# Filter the dataset to show only rows where daily confirmed cases > 10,000.

filtered = d[d["confirmed"] > 10000]
print(filtered)
```

```python
# Group the data by state_unionterritory and compute:

# Total confirmed cases

# Total deaths

# Maximum recovered cases in a single day





d["daily_recovered"] = d.groupby("state_unionterritory")["cured"].diff()



daily_recovered_stats = d.groupby("state_unionterritory").agg(
    total_confirmed=("confirmed", "sum"),
    total_deaths=("deaths", "sum"),
    max_daily_recovered=("daily_recovered", "max")
).sort_values(by="total_confirmed", ascending=False)

print(daily_recovered_stats)
```

```python
# Add a new column death_rate = total_deaths / total_confirmed

d["death_rate"] = d["deaths"] / d['confirmed'] *100
d
```

```python
# Identify the state with the highest death rate.

death_rate_by_state = d.groupby("state_unionterritory").agg(
    total_deaths=("deaths", "sum"),
    total_confirmed=("confirmed", "sum")
)
# print(death_rate_by_state)

death_rate_by_state["death_rate"] = (death_rate_by_state["total_deaths"] / death_rate_by_state["total_confirmed"]) * 100

# Sort to find the highest
highest_death_rate = death_rate_by_state.sort_values(by="death_rate", ascending=False)

print(highest_death_rate.head(1))  # State with highest death rate
```

```python
# Create a column is_high_risk, set to True if the death rate > 2%, else False.
death_rate_by_state["is_high_risk"] = death_rate_by_state["death_rate"] > 2
print(death_rate_by_state)
```

```python
# ðŸ”¢ Top 5 states with the highest confirmed cases?

top_5_states = d.groupby("state_unionterritory").agg(
    total_confirmed=("confirmed", "sum")
).sort_values(by="total_confirmed", ascending=False)

print(top_5_states.head(5))
```

```python
# ðŸ¥ Which state had the best recovery rate (max recovery / confirmed)?

Best_cured_state  = d.groupby("state_unionterritory").agg(
    total_confirmed=("confirmed", "sum"),
    total_cured=("cured", "sum"))
Best_cured_state["recovery_rate"] = Best_cured_state["total_cured"] / Best_cured_state["total_confirmed"]
Best_cured_state.sort_values(by="recovery_rate", ascending=False)
Best_cured_state.head(1)
```

```python
# âš–ï¸ Are there more high-risk or low-risk states based on death rate?

death_rate_state = d.groupby("state_unionterritory").agg(
    total_deaths=("deaths", "sum"),
    total_confirmed=("confirmed", "sum")
)

# Compute death rate %
death_rate_state["death_rate"] = (death_rate_state["total_deaths"] / death_rate_state["total_confirmed"]) * 100

death_rate_state

death_rate_state["risk_level"] = death_rate_state["death_rate"].apply(
    lambda x: "High Risk" if x >= 2 else "Low Risk"
)

death_rate_state
```
