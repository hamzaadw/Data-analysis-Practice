```python
import pandas as pd

data = pd.read_csv("netflix_titles.csv")

print("Columns are ", data.columns )
print("\n")
print(data.info())
```

```python
data.isnull().sum()

Cleaned_data=data.dropna(subset = ["title", "type"])
```

```python
Cleaned_data["date_added"] = pd.to_datetime(Cleaned_data["date_added"].str.strip())
Cleaned_data
```

```python
after2015 = Cleaned_data[Cleaned_data["release_year"]>2015]
after2015
```

```python
# All Movies from country India

Movies_by_india = Cleaned_data[Cleaned_data["country"]=="India"]
```

```python
Cleaned_data["duration_minutes"] = Cleaned_data["duration"].str.extract('(\d+)').astype(float)
Cleaned_data[(Cleaned_data["type"] == "Movie") & (Cleaned_data["duration_minutes"] > 90)]
```

```python
# Reales decade

Cleaned_data["Decade"] = (Cleaned_data['release_year'] //10 )*10
Cleaned_data["Decade"] = Cleaned_data["Decade"].astype(str) +"s"
Cleaned_data
```

```python
print("Unique types:", Cleaned_data["type"].unique())
print("\n")
print("Unique ratings:", Cleaned_data["rating"].unique())
print("\n")
# print("Unique countries:", Cleaned_data["country"].unique())
```

```python
# Group by type and count the number of titles in each category
Cleaned_data.groupby("type").agg(
    count=("title", "count")
)
```

```python
# - Group by country and find the top 5 countries with the most titles

Cleaned_data.groupby("country").agg(
    count = ("title","count")

).sort_values(by="count", ascending=False).head()
```

```python
# - Group by rating and find average release year for each rating type

Cleaned_data.groupby("rating").agg(
    average_release_year = ("release_year", "mean")
).astype(int)
```

```python
# - Group by release_year and show:
# - Total number of titles
# - Count of Movies and TV Shows separately

Cleaned_data.groupby("release_year").agg(
    total_titles = ("title","count"),
    movies = ("type", lambda x: (x == "Movie").sum()),
    tv_shows = ("type",lambda x : (x=="TV Show").sum())
)

```

```python
# - Create a new column `is_recent` True if `release_year` > 2015

Cleaned_data["is_recent"] = Cleaned_data["release_year"] > 2015
Cleaned_data
```

```python
# - Group by country and count how many recent shows are there per country

Cleaned_data.groupby("country").agg(
    recent_shows = ("is_recent", lambda x: (x == True).sum()),

).sort_values(by="recent_shows", ascending=False)
```

```python
# 1. What is the most common rating for TV Shows?

Cleaned_data["rating"].mode()
```

```python
# 2. Which country has the most content on Netflix?

Cleaned_data["country"].mode()
```

```python
# 3. Are most Netflix titles older or recent (after 2015)?

Cleaned_data["release_year"].mode()
print("After 2015",Cleaned_data["release_year"].mode())
```

```python
# 4. Whats the average release year for Netflix Movies?

round(Cleaned_data["release_year"].dropna().mean())
```
